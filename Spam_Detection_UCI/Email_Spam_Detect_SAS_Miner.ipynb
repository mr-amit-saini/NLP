# Email Spam Detection Model using SAS Miner

### Step 1: Import the file using File Import node
- Checck for maximum number of rows to input and change according to need. Default is 1 million.
- Check number of columns to import.
- Right click File Import node and click Edit Variables to change and assign roles to the variables.
- Target Variable is the output variable.

### Step 2: Create a Text Parsing node
- Change POS to No if not required.
- Change Find Entities to Standard.
- Apply Stemming by selecting Stem Terms to Yes.
- Check Edit Variables option same as previous step.

### Step 3: Add Text Filter Node
- Change Check Spelling option to Yes or No as required.
- Select Term Weight as IDF or Entropy.
- After running the node, use Filter Viewer to combine more synonym words and identify the relation between words.
- Select Concept Linking to see the related words to any important word.

### Step 4: Create a Text Topic node
- This is an optional step and has no impact on overall model performance.
- Use this step to see the relation between the words and major topics.

### Step 5: Create a Text Cluster node
- Change SVD Resolution to High.
- Change maximum number of clusters count as required. Default is 40.

### Step 6: Add train test data split using Data Partition node
- Select values for Training, Validation and Test data sets.

### Step 7: Add Metadata Node
- Check Train variables.
- Change the roles of any variable if required.

### Step 8: Apply Decision Tree, Neural Network, or Regression Model Node
- Check Misclassification rate in the report.
- Also check FPR and FNR to determine model accuracy.
